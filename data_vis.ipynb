# Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cs269_winter26_project_data/proteomics_ordered.csv")
display(data.head(10))
```

```python
# average of strand

data['Strain Only'] = data['Strains'].str.split('_').str[0]

averages = data.groupby('Strain Only')['Aldh1a7'].mean()

print(averages)
```

```python
subset = averages.iloc[:10]

subset.plot(kind='bar')
plt.title("Box Plot")
plt.xlabel('Strains')
plt.ylabel('Average Aldh1a7')
plt.show()
```

```python
AJ_subset = data[data['Strain Only'] == 'A.J']
averages = AJ_subset.mean(numeric_only=True)

print(averages)
```

```python
subset = averages.iloc[:10]

subset.plot(kind='bar')
plt.title("Box Plot for Breed A.J")
plt.xlabel('Proteins')
plt.ylabel('Average')
plt.show()
```
